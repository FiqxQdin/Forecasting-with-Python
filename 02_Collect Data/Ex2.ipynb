```python
import matplotlib.pyplot as plt
```

```python
data['Date'] = pd.to_datetime(data['Date'])
```

```python
plt.figure( figsize=(10,6) )
plt.plot(data['Date'],data['Parcel Volume'], marker='o', linestyle='-',color='b',label='Parcel Volume')
plt.xlabel('Date')
plt.ylabel('Parcel Volume')
plt.title('Parcel Volume vs Time')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting all metrics on a single multivariate time series plot
plt.figure(figsize=(12, 8))

# Plot each column
for column in df.columns[1:]:  # Skip the 'Date' column
    plt.plot(data['Date'], data[column], marker='o', linestyle='-', label=column)

# Add labels, title, and legend
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Multivariate Time Series Plot')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create DataFrames for the scales
delivery_times = pd.DataFrame({
    'Rating': range(1, 6),
    'Meaning': [
        'Very Fast',
        'Fast',
        'Average',
        'Slow',
        'Very Slow'
    ]
})

supply_chain_visibility = pd.DataFrame({
    'Rating': range(1, 6),
    'Meaning': [
        'Very Low Visibility',
        'Low Visibility',
        'Moderate Visibility',
        'High Visibility',
        'Very High Visibility'
    ]
})
```

```python
import seaborn as sns
fig, (ax1,ax2) = plt.subplots(2,1, figsize=(10,12))
# Plot for Delivery Times
sns.barplot(x='Rating', y='Meaning', data=delivery_times, ax=ax1, palette='YlOrRd')
ax1.set_title('Delivery Times Scale')
ax1.set_xlabel('Rating')
ax1.set_ylabel('Meaning')

# Plot for Supply Chain Visibility
sns.barplot(x='Rating', y='Meaning', data=supply_chain_visibility, ax=ax2, palette='YlGnBu')
ax2.set_title('Supply Chain Visibility Scale')
ax2.set_xlabel('Rating')
ax2.set_ylabel('Meaning')

plt.tight_layout()
plt.savefig('rating_scales.png')
print("The rating scales have been saved as 'rating_scales.png'")

# Display the DataFrames
print("\
Delivery Times Scale:")
print(delivery_times.to_string(index=False))

print("\
Supply Chain Visibility Scale:")
print(supply_chain_visibility.to_string(index=False))
```
